# Test di plot_utils.py

```python
from mespy import io_utils as io
from mespy import plot_utils as pl
import numpy as np
```

```python
df = io.load_csv("../data/reference/test_misure.csv")

lunghezza = df["lunghezza_mm"].values

pl.histogram(lunghezza)
```

# Esempio di subplots in un unica figure

```python
import matplotlib.pyplot as plt

sigma = df["sigma_mm"].values

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 4), dpi=120, constrained_layout=True)

pl.histogram(lunghezza, bins=8,      xlabel="lunghezza (mm)", ylabel="n misure", title="bins fissi · con stats",  show_mean=True,  show_std=True,  label="campione",   ax=ax1)

pl.histogram(sigma,     bins="sqrt", xlabel="σ (mm)", title="bins sqrt · senza stats", show_mean=False, show_std=False, label="incertezze", ax=ax2, save_path="../figures/histogram_test.png")
```

# Esempio con dati simulati (np.random.normal)

```python
import numpy as np

data_normale = np.random.normal(loc=0, scale=1, size=200)

pl.histogram(
    data_normale,
    bins=20,
    label="Campione simulato",
    xlabel="Valore",
    title="Distribuzione normale simulata",
    save_path="../figures/histogram_normal.png",
)
```

## Tutti i parametri di `histogram` in azione

Mostra `bins`, `ddof`, `label`, `xlabel`, `ylabel`, `title`, `bin_ticks`, `tick_rotation`, `decimals`, `figsize`, `show_legend`, `show_mean`, `show_std`, `dpi`, `xlim`, `ylim` applicati insieme.

```python
x = [2,3,5,1,4,6,4,2,]

pl.histogram(
    x,
    ddof=1,
    bins="auto",
    label="Dati",
    xlabel="Valore",
    ylabel=None,
    title="Istogramma",
    bin_ticks=True,
    tick_rotation=0,
    decimals=3,
    figsize=(8, 5),
    save_path=None,
    show_legend=True,
    show_mean=True,
    show_std=True,
    dpi=300,
    ax=None,
    xlim=None,
    ylim=None,
)

```
