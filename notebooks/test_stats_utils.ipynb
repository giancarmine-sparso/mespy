```python
from mespy import io_utils as io
from mespy import stats_utils as my

import numpy as np
```

```python
df = io.load_csv("../data/reference/test_misure.csv")

misura    = df["misura_n"].values
lunghezza = df["lunghezza_mm"].values
sigma     = df["sigma_mm"].values

df
```

```python
# Statistiche base
media = my.weighted_mean(lunghezza)
var = my.variance(lunghezza)
std = my.standard_deviation(lunghezza)
med = my.median(lunghezza)

w = 1 / np.array(sigma, dtype=float)**2

media_pesata = my.weighted_mean(lunghezza, w)
std_pesata = my.standard_deviation(lunghezza, w)

print(f"Media pesata: {media_pesata:.4f} mm")
print(f"Dev. std pesata: {std_pesata:.4f} mm")
```

```python
cov = my.covariance(lunghezza, misura)
print(f"Covarianza (lunghezza, misura_n): {cov:.6f}")
```
